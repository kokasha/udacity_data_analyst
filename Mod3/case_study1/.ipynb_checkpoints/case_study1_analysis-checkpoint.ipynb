```python
wine_df.rename(columns={'residual sugar':'residual_sugar'},inplace=True)
```

```python
wine_high_sweet = wine_df.query('residual_sugar >= 3.0')
wine_low_sweet = wine_df.query('residual_sugar < 3.0')
```

# Visualtization and communication

### #1: Do wines with higher alcoholic content receive better ratings?

```python
mean_high_alc = wine_high_alc.quality.mean()
mean_low_alc = wine_low_alc.quality.mean()
```

```python
mean_high_alc,mean_low_alc
```

```python
# Create the bar chart

# X locations for the Bar
location = [1,2]
heights = [mean_high_alc,mean_low_alc]
labels = ['High','Low']
plt.bar(location,heights,tick_label=labels)
plt.title('Average Quality Rating for Alcohol')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating')
plt.show()
```

### #2: Do sweeter wines receive higher ratings?

```python
mean_high_sweet = wine_high_sweet.quality.mean()
mean_low_sweet = wine_low_sweet.quality.mean()
mean_high_sweet,mean_low_sweet
```

```python
sweet_heights = [mean_high_sweet,mean_low_sweet]
plt.xlabel('Sweet level')
plt.ylabel('Average Quality Rating')
plt.title('Average Quality Rating based on Sweet Level')
plt.bar(location,sweet_heights,tick_label=labels)
plt.show()
```

### #3: What level of acidity receives the highest average rating?


```python
list(wine_df.groupby('acidity level')['quality'].mean())
```

```python
list(wine_df.groupby('acidity level')['quality'].mean().keys())
```

```python
wine_df.groupby('acidity level')['quality'].mean().keys()
```

```python
mean_acidity_level = list(wine_df.groupby('acidity level')['quality'].mean())
location = [1,2,3,4]
plt.xlabel('Acidity levels')
plt.ylabel('Average Quality Rating')
plt.title('Average Quality Rating based on Acidity Level')
labels = list(wine_df.groupby('acidity level')['quality'].mean().keys())
plt.bar(location,mean_acidity_level,tick_label=labels)
plt.show()
```
